```python
import pandas as pd
```

```python
dataset = pd.read_csv("liuzhou.csv")
```

```python
dataset
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(dataset.AQI指数,dataset.质量指标)
```

```python
x = dataset.AQI指数
y = dataset.质量指标
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_shape=(1,)))#dense 输出维度为1，输入数据维度为1，用元组1，
```

```python
model.summary() #初始化a，b   y=ax+b
```

```python
model.compile(optimizer='adam',
              loss='mse'
)  
```

```python
history = model.fit(x,y,epochs=100)
```

```python
%cd E:/PaddleDetection

# 用 Python 内置库下载（替代 curl，避免下载失败）
import requests

# PP-YOLOv2 官方权重地址
weight_url = "https://bj.bcebos.com/v1/paddledet/models/ppyolo_r50vd_dcn_1x_coco.pdparams"
weight_path = "ppyolo_r50vd_dcn_1x_coco.pdparams"

# 下载权重（显示进度）
response = requests.get(weight_url, stream=True)
total_size = int(response.headers.get('content-length', 0))
downloaded_size = 0

with open(weight_path, 'wb') as f:
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)
            downloaded_size += len(chunk)
            # 打印下载进度
            print(f"下载进度：{downloaded_size/total_size*100:.1f}%", end='\r')

print("\n✅ 权重文件下载完成！")

# 再次验证文件存在
print("当前目录下的权重文件：")
!dir | findstr /i "ppyolo_r50vd_dcn_1x_coco"
```

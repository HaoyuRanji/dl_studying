```python
import tensorflow as tf
print ('TensorFlow Version:{}'.format(tf.__version__))
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(train_image, train_label),(test_image, test_label) = tf.keras.datasets.fashion_mnist.load_data()#从网上搞了很多张im图片
#保存在 C:\Users\雨寂\.keras\datasets\fashion-mnist
```

```python
train_image = train_image/255
test_image = test_image/255 #归一化
```

```python
train_image.shape
```

```python
#每一个层可以被调用，给一个输入，可以给你输出。
input = keras.Input(shape=(28,28))#输入是28*28
```

```python
x = keras.layers.Flatten()(input)  #传进input ，返回x
```

```python
x = keras.layers.Dense(32,activation='relu')(x) #传进x ，返回x
```

```python
x = keras.layers.Dropout(0.5)(x)
```

```python
x = keras.layers.Dense(64,activation='relu')(x)
```

```python
output = keras.layers.Dense(10,activation='softmax')(x)
```

```python
model = keras.Model(inputs=input, outputs = output)
```

```python
model.summary()  #建立好模型
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['acc']
)  #每次运行计算正确率 shift+tap可以看函数默认值参数

history = model.fit(train_image,train_label,
          epochs=5,
          validation_data=(test_image,test_label)
         )#用验证集来检验模型
```
